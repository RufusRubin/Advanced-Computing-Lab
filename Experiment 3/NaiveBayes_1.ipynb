**Load the dataset**

```python
from sklearn.datasets import load_wine
data = load_wine()
```

```python
x= data.data
```

```python
y=data.target
```

**train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

**Import model**

```python
from sklearn.naive_bayes import GaussianNB
model_1=GaussianNB()
from sklearn.naive_bayes import MultinomialNB
model_2=MultinomialNB()
from sklearn.naive_bayes import BernoulliNB
model_3=BernoulliNB(binarize=2)
```

**Fit the model**

```python
model_1= model_1.fit(X_train,y_train)
model_2= model_2.fit(X_train,y_train)
model_3= model_3.fit(X_train,y_train)
```

**predict**

```python
res_1= model_1.predict(X_test)
res_2= model_2.predict(X_test)
res_3= model_3.predict(X_test)
```

**Acccuracy**

```python
from sklearn.metrics import accuracy_score
score_1= accuracy_score(res_1, y_test)
print("Gaussian Naive Bayes:",score_1*100)
```

```python
from sklearn.metrics import accuracy_score
score_2= accuracy_score(y_test,res_2)
print("Multinomial Naive Bayes:",score_2*100)
```

```python
from sklearn.metrics import accuracy_score
score_3= accuracy_score(res_3, y_test)
print("Bernoulli Naive Bayes:",score_3*100)
```
