```python
#load dataset
import pandas as pd
```

```python
data= pd.read_csv("/content/IRIS.csv")
```

```python
x=data[["sepal_length","sepal_width","petal_length","petal_width"]][0:50]
```

```python
y=["species"]
```

```python
#apply PCA
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca=pca.fit_transform(x)
```

```python
new_data = pd.DataFrame(pca,columns=['PC1' ,'PC2'])
new_data
```

```python
 x_axis= new_data['PC1']
```

```python
y_axis = new_data['PC2']
```

```python
#plot the figure 
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_axis,y_axis)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
#hierarchical clustering 
from scipy.cluster import hierarchy 
```

```python
a=hierarchy.linkage(x,"complete")
```

```python
import matplotlib.pyplot as plt
hierarchy.dendrogram(a,orientation="top",above_threshold_color="green")
plt.figure()
```
