**Load the dataset**

```python
import pandas as pd
```

```python
data= pd.read_csv('/content/Wheat_seed.csv')
```

```python
x=data[['area','perimeter','compactness'	,'lengthOfKernel',	'widthOfKernel'	,'asymmetryCoefficient'	,'lengthOfKernelGroove']]
```

```python
y=data['seedType']
```

**train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

**import th model**

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel = 'rbf')
```

**Fit the model**


```python

model = model.fit(X_train,y_train)
```

```python

res = model.predict(X_test)
```

**Acccuracy**

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(res , y_test)
```

```python
print(score*100)
```
