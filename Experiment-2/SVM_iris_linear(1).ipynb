**Load the dataset**

```python
import pandas as pd
```

```python
data = pd.read_csv("/content/iris.csv")
```

```python
x= data[["sepal_length"	,"sepal_width" ,"petal_length","petal_width"]]
y= data["species"]
```

**Import SVM model**

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

**Fit the model**

```python
model=model.fit(x,y)
```

**Predict the model**

```python
res = model.predict([[5.6,6.5,4.3,5]])
```
