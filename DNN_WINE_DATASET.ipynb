```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
x=wine.data
```

```python
y=wine.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
ML = keras.models.Sequential()
```

```python
ML.add(keras.layers.Dense(units=20,activation='relu', input_dim=x.shape[1]))
```

```python
ML.add(keras.layers.Dense(units=10, activation='relu'))
```

```python
ML.add(keras.layers.Dense(units=3,activation='sigmoid'))
```

```python
ML.summary()
```

```python
ML.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
```

```python
ML.fit(x_train,y_train,epochs=50)
test_loss,test_accuracy=ML.evaluate(x_test,y_test)
print(test_loss,test_accuracy)
```

```python
ML.compile(loss='sparse_categorical_crossentropy', optimizer='SGD',metrics=['accuracy'])
```

```python
ML.fit(x_train,y_train,epochs=50)
test_loss,test_accuracy=ML.evaluate(x_test,y_test)
print(test_loss,test_accuracy)
```

```python
ML.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop',metrics=['accuracy'])
```

```python
ML.fit(x_train,y_train,epochs=50)
test_loss,test_accuracy=ML.evaluate(x_test,y_test)
print(test_loss,test_accuracy)
```

```python
ML.compile(loss='sparse_categorical_crossentropy', optimizer='adagrad',metrics=['accuracy'])
ML.fit(x_train,y_train,epochs=50)
test_loss,test_accuracy=ML.evaluate(x_test,y_test)
print(test_loss,test_accuracy)
```

```python
ML.compile(loss='sparse_categorical_crossentropy', optimizer='adadelta',metrics=['accuracy'])
ML.fit(x_train,y_train,epochs=50)
test_loss,test_accuracy=ML.evaluate(x_test,y_test)
print(test_loss,test_accuracy)
```
