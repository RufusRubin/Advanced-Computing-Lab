Load dataset

```python
import pandas as pd
```

```python
data= pd.read_csv("/content/iris.csv")
```

```python
x= data[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
y= data['species']
```

apply pca

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
pca=pca.fit_transform(x)
```

```python
new_data = pd.DataFrame(pca,columns=['PC1' ,'PC2'])
new_data
```

```python
 x_axis= new_data['PC1']
```

```python
y_axis = new_data['PC2']
```

PLOT the figure

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_axis,y_axis)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model= KMeans(n_clusters=2, max_iter=100 , random_state=0)
```

```python
model=model.fit(new_data)
```

```python
centres = model.cluster_centers_
```

```python
centres
```

```python
labels = model.labels_
```

```python
labels
```

```python
plt.scatter(x_axis ,y_axis,c=labels,cmap='rainbow')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.scatter(centres[0][0],centres[0][1],color='orange')
plt.scatter(centres[0][1],centres[1][1],color='green')
```
